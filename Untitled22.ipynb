```python
!pip install kmodes

import pandas as pd
import numpy as np
df = pd.read_csv('marketing_campaign.csv', sep='\t')

# Aperçu des premières lignes et de la structure des données
print(df.shape)
display(df.head())
df.info()
```

```python
# Suppression des colonnes non utiles
cols_drop = ['ID','Z_CostContact','Z_Revenue','Dt_Customer',
             'AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5',
             'Complain','Response']
df = df.drop(columns=cols_drop, errors='ignore')

# Remplacement des NaN d'Income par la médiane
df['Income'] = df['Income'].fillna(df['Income'].median())

# Suppression des outliers :
df = df[df['Income'] < 100000]  # supprime l'énorme 666666
df = df[df['Year_Birth'] >= 1930]  # exclut les naissances avant 1930
# Exclusion des catégories aberrantes de statut marital
df = df[~df['Marital_Status'].isin(['Absurd','YOLO','Alone'])]

# Création de la colonne Age et suppression de Year_Birth
df['Age'] = 2025 - df['Year_Birth']
df = df.drop(columns=['Year_Birth'])

# Vérification finale après nettoyage
print(df.shape)
df.isna().sum()  # plus de valeurs manquantes
```

```python
# Liste des colonnes sélectionnées pour la segmentation
cols_consumption = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
cols_behavior = ['Recency','NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']
cols_demo = ['Income','Age','Education','Marital_Status','Kidhome','Teenhome']

cols_selected = cols_consumption + cols_behavior + cols_demo
df_seg = df[cols_selected].copy()
df_seg.head()
```

```python
# Séparation des données numériques et catégorielles
num_cols = ['Income','Kidhome','Teenhome','Age'] + cols_consumption + cols_behavior
cat_cols = ['Education','Marital_Status']
print("Vérification des NaN dans df_seg avant le split :")
print(df_seg.isna().sum())
for col in num_cols:
    if df_seg[col].isna().any():
        df_seg[col] = df_seg[col].fillna(df_seg[col].median())

for col in cat_cols:
    if df_seg[col].isna().any():
        print(f"Filling NaN in categorical column {col} with 'Missing'.")
        df_seg[col] = df_seg[col].fillna('Missing') # Use a placeholder like 'Missing'

print("\nVerification des NaN dans df_seg après imputation :")
print(df_seg[cols_selected].isna().sum())
# Standardisation des variables numériques
scaler = StandardScaler()
df_num = pd.DataFrame(scaler.fit_transform(df_seg[num_cols]), columns=num_cols)

# Codage numérique des variables catégorielles
df_cat = df_seg[cat_cols].copy()
for col in cat_cols:
    # Assurez-vous qu'il n'y a pas de NaN ici avant de convertir en 'category'
    # Si df_seg[col].isna().any() est True ici, cela causera problème.
    df_cat[col] = df_cat[col].astype('category').cat.codes

# Matrice finale de caractéristiques pour K-Prototypes
X = pd.concat([df_num, df_cat], axis=1)
print("\nVerification finale des NaN dans la matrice X :")
print(X.isna().sum().sum())
# Indices des colonnes catégorielles dans la matrice X
cat_indices = [X.shape[1] - len(cat_cols) + i for i in range(len(cat_cols))]
print("Colonnes catégorielles aux indices :", cat_indices)
X_clean = X.dropna()
cat_indices = [X_clean.shape[1] - len(cat_cols) + i for i in range(len(cat_cols))]
print("Colonnes catégorielles aux indices :", cat_indices)
print("\nVerification finale des NaN dans la matrice X :")
print(X_clean.isna().sum().sum())
```

```python
# Création de X_all avec vérification des NaN
X_all = pd.concat([df_num, pd.get_dummies(df_seg[cat_cols])], axis=1)

# Vérification des NaN
print("NaN dans df_num:", df_num.isna().sum().sum())
print("NaN dans les dummies:", pd.get_dummies(df_seg[cat_cols]).isna().sum().sum())

# Suppression des lignes avec NaN (doit être les mêmes que pour X_clean)
X_all = X_all.dropna()

# Conversion en numpy array seulement après nettoyage
X_all_values = X_all.astype(float).values

# Vérification finale
print("NaN dans X_all après nettoyage:", np.isnan(X_all_values).any())
```

```python
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score, davies_bouldin_score

costs = []
sil_scores = []
db_scores = []
K_range = range(2, 9)
for k in K_range:
    # Vérification de X_clean (déjà fait précédemment)

    kproto = KPrototypes(n_clusters=k, init='Cao', n_init=5, random_state=42)
    clusters = kproto.fit_predict(X_clean, categorical=cat_indices)
    costs.append(kproto.cost_)

    # Calcul des métriques
    try:
        sil = silhouette_score(X_all_values, clusters, metric='euclidean')
        dbi = davies_bouldin_score(X_all_values, clusters)
    except Exception as e:
        print(f"Erreur avec k={k}: {str(e)}")
        sil, dbi = np.nan, np.nan

    sil_scores.append(sil)
    db_scores.append(dbi)
    print(f"k={k}: coût={kproto.cost_:.1f}, Silhouette={sil:.3f}, Davies-Bouldin={dbi:.3f}")
```

```python
import matplotlib.pyplot as plt

# Graphe Elbow (coût) et Silhouette
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(list(K_range), costs, marker='o')
plt.title("Méthode du coude (coût K-Prototypes)")
plt.xlabel("Nombre de clusters k")
plt.ylabel("Coût total du clustering")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(list(K_range), sil_scores, marker='o', color='green', label='Silhouette')
plt.plot(list(K_range), db_scores, marker='o', color='red', label='Davies-Bouldin')
plt.title("Scores de silhouette et Davies-Bouldin")
plt.xlabel("Nombre de clusters k")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Clustering final avec K-Prototypes
k_opt = 4
kproto = KPrototypes(n_clusters=k_opt, init='Cao', n_init=5, verbose=1, random_state=42)
# Perform clustering on X_clean, which has the same index as X_all after dropna()
clusters = kproto.fit_predict(X_clean, categorical=cat_indices)

# Assign clusters back to df_seg using the index of X_clean to align them correctly
# This ensures that the cluster assignments are only added to the rows that were included in the clustering
df_seg.loc[X_clean.index, 'Cluster'] = clusters

# Répartition du nombre de clients par cluster
# Use .value_counts() on the cluster column, which now might have NaNs for the rows that were dropped
print(df_seg['Cluster'].value_counts(dropna=False).sort_index())
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Ensure matplotlib is imported

# The issue is that X_num_pca was computed on df_num before dropping NaNs.
# The 'clusters' variable corresponds to the rows that remained after dropping NaNs in X_clean.
# We need to align the PCA data with the clusters by performing PCA on the numerical part
# of the data *after* dropping the NaNs.

# Re-extract the numerical part of the cleaned data (X_clean)
# X_clean contains both numerical and categorical columns.
# The first len(num_cols) columns in X_clean are the numerical ones.
X_num_cleaned = X_clean[num_cols] # Select only numerical columns from X_clean

# Perform PCA on the numerical data that corresponds to the clustered rows
X_num_pca = PCA(n_components=2).fit_transform(X_num_cleaned.values)

plt.figure(figsize=(6,6))
# Now X_num_pca and clusters should have the same number of rows
sns.scatterplot(x=X_num_pca[:,0], y=X_num_pca[:,1], hue=clusters, palette='Set2', legend='full')
plt.title("Projection PCA des clients (2D, couleur = cluster)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Cluster")
plt.show()
```
